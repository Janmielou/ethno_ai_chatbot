```python
# Page counts
total_pages_per_manuscript = 150
color_pages_per_manuscript = 15
bw_pages_per_manuscript = total_pages_per_manuscript - color_pages_per_manuscript

# Prices per page
price_per_bw_page = 3      # non-colored
price_per_color_page = 5   # colored

# Number of copies/manuscripts to print
num_manuscripts = 4

# Cost calculations
cost_per_manuscript = (
    bw_pages_per_manuscript * price_per_bw_page
    + color_pages_per_manuscript * price_per_color_page
)

total_cost = cost_per_manuscript * num_manuscripts

print(f"Estimated cost of {total_pages_per_manuscript} pages (per manuscript): ₱{cost_per_manuscript:,.2f}")
print(f"Total cost for {num_manuscripts} manuscripts: ₱{total_cost:,.2f}")
```
